```python
import pandas as pd
import numpy as np
from datetime import datetime as dt
import re
import matplotlib
```

```python
Jan = pd.read_csv('Sales_January_2019.csv')
Feb = pd.read_csv('Sales_February_2019.csv')
Mar = pd.read_csv('Sales_March_2019.csv')
Apr = pd.read_csv('Sales_April_2019.csv')
May = pd.read_csv('Sales_May_2019.csv')
Jun = pd.read_csv('Sales_June_2019.csv')
Jul = pd.read_csv('Sales_July_2019.csv')
Aug = pd.read_csv('Sales_August_2019.csv')
Sep = pd.read_csv('Sales_September_2019.csv')
Oct = pd.read_csv('Sales_October_2019.csv')
Nov = pd.read_csv('Sales_November_2019.csv')
Dec = pd.read_csv('Sales_December_2019.csv')
```

```python
Jan_Dec = [Jan, Feb, Mar, Apr, May, Jun, Jul, Aug, Sep, Oct, Nov, Dec]
raw_data = pd.concat(Jan_Dec)
#raw_data.info()
```

```python
null_values = raw_data[(raw_data.isnull()).any(axis = 1)]
#null_values.info()
```

```python
raw_data.dropna(subset = ["Order ID"], inplace=True)
#raw_data.info()
raw_data.head()
```

```python
updated_data = raw_data[~ raw_data['Order ID'].str.contains('Order ID')]
updated_data.head()
```

```python
updated_data[['Date', 'Time']] = updated_data['Order Date'].str.split(' ', expand = True)
updated_data.head()
```

```python
updated_data["Date"] = updated_data["Date"].apply(lambda a : dt.strptime(a, "%d/%m/%y"))
```

```python
updated_data[['Month', 'Day', 'Year']] = updated_data['Date'].str.split('/', expand = True)
updated_data.head()
```

```python
updated_data['Year'].replace({'2019': '19'}, inplace=True)
updated_data
```

```python
#updated_data['Quantity Ordered'] = pd.to_numeric(raw_data['Quantity Ordered'])
#updated_data['Price Each'] = pd.to_numeric(raw_data['Price Each'])
updated_data['Amount'] = updated_data['Quantity Ordered'].astype(int) * updated_data['Price Each'].astype(float)
updated_data
```

```python
updated_data = updated_data.drop(columns = ['Date'])
updated_data.head()
```

```python
cols = list(updated_data.columns)
updated_data = updated_data[cols[0:4] + [cols[10]] + cols[7:10] + [cols[6]] + cols[4:6]]
updated_data.head()
```

```python
updated_data['Year'] = updated_data['Year'].replace('19','2019')
```

```python
updated_data['Date'] = updated_data['Year'].astype(str) + updated_data['Month'].astype(str) + updated_data['Day'].astype(str)
updated_data['Date'] = pd.to_datetime(updated_data["Date"], format = '%Y%m%d')
updated_data
```

```python
updated_data["Date"] = updated_data["Date"].apply(lambda a : dt.strptime(a, "%Y/%m/%d"))
```

```python
updated_data['Month'] = updated_data['Date'].dt.month_name()
updated_data.head()
```
